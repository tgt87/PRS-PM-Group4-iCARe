## Public Transport (Date format: YYYY)

```python
df_public_transport = pd.read_csv('yyyy/Public-Transport.csv')
df_public_transport = df_public_transport.replace('na',np.NaN)
df_public_transport = df_public_transport.replace('-',np.NaN)
df_public_transport['date'] = df_public_transport['Year'].map(lambda x: str(int(x))+'-12-31')
df_public_transport['date'] = pd.to_datetime(df_public_transport['date'])
df_public_transport.drop('Year', axis=1, inplace=True)
df_public_transport.head()
```

## Household Net Worth (Date format:YYYY-Q)

```python
df_household_net_worth = pd.read_csv('yyyy-Q/Household-Net-Worth.csv')
df_household_net_worth
```

```python
def Q_generate_date(y,q):
    year = str(y)
    if q == 1:
        month = '01'
    elif q == 2:
        month = '04'
    elif q == 3:
        month = '07'
    else:
        month = '10'
    return year+'-'+month+'-01'

df_household_net_worth['date'] = df_household_net_worth.apply(lambda x: Q_generate_date(x.Year,x.Quarter), axis=1)
df_household_net_worth['date'] = pd.to_datetime(df_household_net_worth['date']).dt.to_period('M')
df_household_net_worth.drop(['Year','Quarter'], axis=1, inplace=True)
```

```python
df_pivoted = df_household_net_worth.set_index('date')
df_pivoted = df_pivoted.resample('M').interpolate() #convert Quarterly to Monthly 
#df_pivoted = df_pivoted.resample('SMS').mean().bfill() #SMS - 1 & 15 of month
df_pivoted.to_csv('household_net_worth_new.csv')
df_pivoted.head()
```

## Domestic Interest Rate (Date format:YYYY-MM-DD)

```python
df_domestic_interest_rates = pd.read_csv('yyyy-mm-dd/Domestic-Interest-Rates.csv',skiprows=6)
df_domestic_interest_rates = df_domestic_interest_rates.iloc[:9883]
df_domestic_interest_rates = df_domestic_interest_rates.replace('na',np.NaN).replace('-',np.NaN)
df_domestic_interest_rates = df_domestic_interest_rates.rename(columns = dict(zip(df_domestic_interest_rates.columns[[0,1,2]],['Year','Month','Day'])))

#df_domestic_interest_rates[pd.to_numeric(df_domestic_interest_rates['Year'],errors='coerce')]
df_domestic_interest_rates = df_domestic_interest_rates[pd.to_numeric(df_domestic_interest_rates['Day'],errors='coerce').notnull()]

for i in df_domestic_interest_rates.columns[0:2]:
    df_domestic_interest_rates[i].fillna(method='ffill',inplace=True)
for j in df_domestic_interest_rates.columns:
    df_domestic_interest_rates[j].fillna(method='ffill',inplace=True)


df_domestic_interest_rates.dropna(subset=['Day'],inplace=True)
```

```python
df_domestic_interest_rates
```

```python
month_mapping = {
    "Jan": "01",
    "Feb": "02",
    "Mar": "03",
    "Apr": "04",
    "May": "05",
    "Jun": "06",
    "Jul": "07",
    "Aug": "08",
    "Sep": "09",
    "Oct": "10",
    "Nov": "11",
    "Dec": "12"
}

def generate_date(y,m,d):
    year = str(y)
    month = month_mapping[m]
    day = str(int(d))
    #print(year+'-'+month+'-'+day)
    return year+'-'+month+'-'+day

df_domestic_interest_rates['date'] = df_domestic_interest_rates.apply(lambda x: generate_date(x.Year,x.Month,x.Day), axis=1)

df_domestic_interest_rates = df_domestic_interest_rates.loc[df_domestic_interest_rates['Day'].isin([1,15])] #filter 1 & 15 of month

df_domestic_interest_rates['date'] = pd.to_datetime(df_domestic_interest_rates['date'])  
df_domestic_interest_rates.drop(['Year','Month','Day'], axis=1, inplace=True)
    
df_domestic_interest_rates.head()
```

```python
df_pivoted = df_domestic_interest_rates.set_index('date')
df_pivoted.to_csv('domestic_interest_rates_new.csv')
df_pivoted.head()
```

## Personal Disposable Income

```python
df_disposable_income = pd.read_csv('Personal-Disposable-Income.csv')
df_disposable_income = df_disposable_income.set_index('Data Series').T
df_disposable_income.reset_index(inplace=True)
df_disposable_income.head()
```

```python
import calendar
import datetime

quarter_mapping = {
    "1Q": 3,
    "2Q": 6,
    "3Q": 9,
    "4Q": 12
}

def convert_quarter(quarter):
    splitted = quarter.split(' ')
    year = int(splitted[0])
    month = quarter_mapping[splitted[1]]
    last_day_of_month = datetime.date(year, month, calendar.monthrange(year, month)[1])
    return last_day_of_month
```

```python
df_disposable_income['date'] = df_disposable_income['index'].map(convert_quarter)
df_disposable_income['date'] = pd.to_datetime(df_disposable_income['date'])
df_disposable_income.drop('index', axis=1, inplace=True)
df_disposable_income.set_index('date', inplace=True)
df_disposable_income.head()
```

```python
df_disposable_income = df_disposable_income.resample('SMS').mean().bfill()
df_disposable_income.to_csv('personal_disposable_income_new.csv')
df_disposable_income.head()
```
