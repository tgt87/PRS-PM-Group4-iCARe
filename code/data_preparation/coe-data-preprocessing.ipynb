### CPI Dataset

```python
cpi_dataset = 'Consumer-Price-Index-Clean.csv'
cpi_df = pd.read_csv(dataset_folderpath+cpi_dataset, index_col='Data Series')
print('CPI dataset')
cpi_df.head()
```

```python
## tranpose CPI dataset so that date becomes the index column
cpi_df_t = cpi_df.transpose().reset_index()
cpi_df_t.head()
```

```python
print('CPI dataset latest date:', cpi_df_t.loc[0, 'index'])
print('CPI dataset oldest date:', cpi_df_t.loc[cpi_df_t.shape[0]-1, 'index'])
```

```python
from pandas.tseries.offsets import MonthEnd
cpi_df_t['year'] = [x[:4] for x in cpi_df_t['index']]
cpi_df_t['month'] = [x[5:] for x in cpi_df_t['index']]
cpi_df_t['month'] = cpi_df_t['month'].str.strip()
cpi_df_t['day'] = 1

month_dict = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
for i in range(cpi_df_t.shape[0]):
    cpi_df_t.loc[i, 'month'] = month_dict[cpi_df_t.loc[i, 'month']]
    
cpi_df_t['Date'] = pd.to_datetime(cpi_df_t[['year', 'month', 'day']])
cpi_df_t['Date'] = pd.to_datetime(cpi_df_t['Date'], format="%Y-%m-%d") + MonthEnd(1)
cpi_df_t.set_index('Date', inplace=True)
cpi_df_t.drop(columns=['year', 'month', 'day', 'index'], inplace=True)
                
```

### Household-Net-Worth

```python
hh_net_worth_dataset = 'Household-Net-Worth-data.csv'
hh_net_worth_df = pd.read_csv(dataset_folderpath+hh_net_worth_dataset)
hh_net_worth_df.head()
```

```python
hh_net_worth_df = hh_net_worth_df.set_index('Data Series').T
hh_net_worth_df.reset_index(inplace=True)
hh_net_worth_df.head()
```

```python
import calendar
import datetime

quarter_mapping = {
    "1Q": 3,
    "2Q": 6,
    "3Q": 9,
    "4Q": 12
}

def convert_quarter(quarter):
    splitted = quarter.split(' ')
    year = int(splitted[0])
    month = quarter_mapping[splitted[1]]
    last_day_of_month = datetime.date(year, month, calendar.monthrange(year, month)[1])
    return last_day_of_month
```

```python
hh_net_worth_df['date'] = hh_net_worth_df['index'].map(convert_quarter)
hh_net_worth_df['date'] = pd.to_datetime(hh_net_worth_df['date'])
hh_net_worth_df.drop('index', axis=1, inplace=True)
hh_net_worth_df.set_index('date', inplace=True)
hh_net_worth_df.head()
```

```python
hh_net_worth_df_resampled = hh_net_worth_df.resample('SMS').mean().bfill()
hh_net_worth_df_resampled.head()
```

```python
hh_net_worth_df_resampled.to_csv('Household-Net-Worth-Clean.csv')
```
